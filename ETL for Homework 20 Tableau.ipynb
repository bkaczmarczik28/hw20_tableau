```python
# import dependecy for pandas
import pandas as pd
```

```python
# Load latest dataset
data201912=pd.read_csv("JC-201912-citibike-tripdata.csv")
data201912.head()
print("Number of rows 201912:" + str(data201912.shape))
```

```python
# load second data set
data201906=pd.read_csv("JC-201906-citibike-tripdata.csv")
data201906.head()
print("Number of rows 201906:" + str(data201906.shape))
```

```python
# combine the two first sets of data
data_full=data201912.append(data201906, ignore_index=True)
print("Number of rows of data_full:" + str(data_full.shape))
```

```python
# Check the appearance of the rows at the end of the appended table
data_full.tail()
```
